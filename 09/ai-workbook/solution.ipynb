## The adversarial bug

[../adversarial_raytrace_buggy.py](./adversarial_raytrace_buggy.py) solves the
quadratic for the ray–sphere intersection but keeps the wrong root:

```python
t = -b + sq        # far root; the nearest hit is -b - sq
```

A ray that pierces a sphere has two intersection distances. The **nearest**
positive one (`-b - √disc`, when positive) is the front surface you actually see;
`-b + √disc` is the exit point on the back. Rendering with the far root draws a
sphere-shaped region (so a thumbnail looks plausible) but every depth is the back
of the sphere, which corrupts the surface normal, shading, and any shadow rays
cast from the hit point.

## The fix

Pick the smallest positive root:

```python
t_near = -b - sq
t_far  = -b + sq
t = np.where(t_near > 1e-6, t_near, t_far)   # nearest positive hit
```

Also **normalise the ray direction** so `t` is a real distance — the code does
this with `dirs / norm(dirs)`. The corrected version is in
[raytrace_solution.py](solutions/raytrace_solution.py).

## Why the verification matters

"It renders a sphere" is not a correctness test — the silhouette is the same for
both roots. The reliable test is a **per-ray comparison of the hit distance `t`**
against a reference, which is what [../verify_raytrace.py](./verify_raytrace.py)
does. It also checks that hit/miss classification matches.

## Mapping to the 5-phase loop

- **PREDICT:** flag the root choice (and the normalisation assumption) from
  reading the code.
- **VERIFY:** compare `t` per ray, not the rendered silhouette.
- **DIAGNOSE:** on a GPU/RT core this is a per-ray kernel; insist a generated
  version passes the `t` comparison before trusting any rendered image.
